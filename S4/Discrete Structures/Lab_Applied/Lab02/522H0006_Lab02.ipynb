```python
#Exercise02

def lLImplies(P, Q):
    R = []
    for i in range(len(P)):
        if P[i] and not Q[i]:
            R.append(False)
        else:
            R.append(True)
    return R

def lLAnd(P, Q):
    R = []
    for i in range(len(P)):
        if P[i] and Q[i]:
            R.append(True)
        else:
            R.append(False)
    return R

def lLOr(P, Q):
    R = []
    for i in range(len(P)):
        if P[i] or Q[i]:
            R.append(True)
        else:
            R.append(False)
    return R

def lLXor(P, Q):
    R = []
    for i in range(len(P)):
        if (P[i] and not Q[i]) or (not P[i] and Q[i]):
            R.append(True)
        else:
            R.append(False)
    return R

def lLNot(P):
    R = []
    for i in range(len(P)):
        R.append(not P[i])
    return R

def lLEquivalent(P, Q):
    R = []
    for i in range(len(P)):
        if P[i] == Q[i]:
            R.append(True)
        else:
            R.append(False)
    return R

P = [True, True, False, False]
Q = [True, False, True, False]

print("P:", P)
print("Q:", Q)
print()
print("a) lLImplies(P, Q):", lLImplies(P, Q))
print("b) lLAnd(P, Q):", lLAnd(P, Q))
print("c) lLOr(P, Q):", lLOr(P, Q))
print("d) lLXor(P, Q):", lLXor(P, Q))
print("e) lLNot(P):", lLNot(P))
print("f) lLEquivalent(P, Q):", lLEquivalent(P, Q))
```

```python
#Exercise04

from itertools import product

# p ∨ q → p ∧ q → ¬p ∨ ¬q
def expression1(p, q):
    return (p or q) <= ((p and q) <= (not p or not q))
print("p | q | Result")
for p, q in product([True, False], repeat=2):
    print(f"{p} | {q} | {expression1(p, q)}")

print()
# ¬p ∨ (q ∧ r) → r
def expression2(p, q, r):
    return (not p or (q and r)) <= r
print("p | q | r | Result")
for p, q, r in product([True, False], repeat=3):
    print(f"{p} | {q} | {r} | {expression2(p, q, r)}")

print()
# (p → q) ∧ (q → r)
def expression3(p, q, r):
    return (not p or q) and (not q or r)
print("p | q | r | Result")
for p, q, r in product([True, False], repeat=3):
    print(f"{p} | {q} | {r} | {expression3(p, q, r)}")

print()
# (p ∨ (q ∧ r)) ↔ ((p ∨ q) ∧ (p ∨ r))
def expression4(p, q, r):
    return (p or (q and r)) == ((p or q) and (p or r))
print("p | q | r | Result")
for p, q, r in product([True, False], repeat=3):
    print(f"{p} | {q} | {r} | {expression4(p, q, r)}")

print()
# p ∨ q → ¬r ∨ t
def expression5(p, q, r, t):
    return (p or q) <= (not r or t)
print("p | q | r | t | Result")
for p, q, r, t in product([True, False], repeat=4):
    print(f"{p} | {q} | {r} | {t} | {expression5(p, q, r, t)}")

print()
# p ∨ t → q → (r → t)
def expression6(p, q, r, t):
    return (p or t) <= (q <= (r <= t))
print("p | q | r | t | Result")
for p, q, r, t in product([True, False], repeat=4):
    print(f"{p} | {q} | {r} | {t} | {expression6(p, q, r, t)}")

print()
# (p ∨ (q ∧ r)) ↔ (((p ∨ q) ∧ (p ∨ r)) ∧ (t ∨ ¬t))
def expression7(p, q, r, t):
    return (p or (q and r)) == (((p or q) and (p or r)) and (t or not t))
print("p | q | r | t | Result")
for p, q, r, t in product([True, False], repeat=4):
    print(f"{p} | {q} | {r} | {t} | {expression7(p, q, r, t)}")
```

```python
#Exercise05

def is_equivalent_1():
    # p ≡ ¬(¬p)
    for p in [True, False]:
        left_side = p
        right_side = not(not p)
        if left_side != right_side:
            return "Inequivalent"
    return "Equivalent"

def is_equivalent_2():
    # ¬(¬q ∧ p) ∧ (q ∨ p) ≡ q
    for p in [True, False]:
        for q in [True, False]:
            left_side = not(not q and p) and (q or p)
            right_side = q
            if left_side != right_side:
                return "Inequivalent"
    return "Equivalent"

def is_equivalent_3():
    # ¬(p ∨ q) ≡ (¬p ∨ ¬q)
    for p in [True, False]:
        for q in [True, False]:
            left_side = not(p or q)
            right_side = (not p) or (not q)
            if left_side != right_side:
                return "Inequivalent"
    return "Equivalent"

def is_equivalent_4():
    # (p ∨ q) → r ≡ (p → r) ∧ (q → r)
    for p in [True, False]:
        for q in [True, False]:
            for r in [True, False]:
                left_side = (p or q) <= r
                right_side = (p <= r) and (q <= r)
                if left_side != right_side:
                    return "Inequivalent"
    return "Equivalent"

def is_equivalent_5():
    # ¬(p ∧ q) ≡ (¬p ∧ ¬q)
    for p in [True, False]:
        for q in [True, False]:
            left_side = not(p and q)
            right_side = (not p) or (not q)
            if left_side != right_side:
                return "Inequivalent"
    return "Equivalent"

def is_equivalent_6():
    # (p ∨ ¬q) → ¬p ≡ (p ∨ (¬q)) → ¬p
    for p in [True, False]:
        for q in [True, False]:
            left_side = (p or (not q)) <= (not p)
            right_side = (p or (not q)) <= (not p)
            if left_side != right_side:
                return "Inequivalent"
    return "Equivalent"

def is_equivalent_7():
    # ¬(p ∨ q) ≡ (¬p ∧ ¬q)
    for p in [True, False]:
        for q in [True, False]:
            left_side = not(p or q)
            right_side = (not p) and (not q)
            if left_side != right_side:
                return "Inequivalent"
    return "Equivalent"

# Test the equivalence of each formula
print("Formula 1:", is_equivalent_1())
print("Formula 2:", is_equivalent_2())
print("Formula 3:", is_equivalent_3())
print("Formula 4:", is_equivalent_4())
print("Formula 5:", is_equivalent_5())
print("Formula 6:", is_equivalent_6())
print("Formula 7:", is_equivalent_7())
```
