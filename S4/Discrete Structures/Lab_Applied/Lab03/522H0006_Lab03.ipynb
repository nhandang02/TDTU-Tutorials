```python
#Exercise02
def formalConvert(S):
    S = S.strip()
    if S.startswith('For all '):
        D = S[8:S.index(',')]
        P = S[S.index(',')+6:]
        F = 'Formal form: For all x in D, P(x)'
    elif S.startswith('Exist a '):
        D = S[8:S.index(',')]
        P = S[S.index(',')+6:]
        F = 'Formal form: Exist x in D, P(x)'
    elif S.startswith('Exist an '):
        D = S[9:S.index(',')]
        P = S[S.index(',')+6:]
        F = 'Formal form: Exist x in D, P(x)'
    elif S.startswith('There is at least one '):
        D = S[21:S.index(',')]
        P = S[S.index(',')+5:]
        F = 'Formal form: Exist x in D, P(x)'
    elif S.__contains__('did not'):
        arr = S.split(' did not ')
        D = arr[0].replace('Every ', '')
        P = "did not " + arr[1]
        F = 'Formal form: For all x in D, P(x)'
    else:
        D = "Invalid Statement"
        P = "Invalid Statement"
        F = "Invalid Statement"

    return [D, P, F]

S = ['For all fishes, they need water to survive',
     'Exist a person, who is left handed',
     'Exist an employee in the company, who is late to work everyday',
     'For all fishes in this pond, they are Koi fish',
     'There is at least one creature in the ocean, it can live on land',
     'Every students in class A did not pass the test',
     'My name is Nhan']

for s in S:
    print(formalConvert(s))
```

```python
#Exercise04
def formalConvertPQ(S):
  S = S.strip()

  if S.startswith('For all '):
    if S.__contains__('then'):
      arr = S.split(' then ')
      D = S[8:S.index(',')]
      P = arr[0][S.index(',')+10:]
      Q = arr[1][5:]
      F = f'Formal form: For all x in D, if P(x) then Q(x)'
  elif S.startswith('For every '):
    if S.__contains__('then'):
      arr = S.split(' then ')
      D = S[10:S.index(',')]
      P = arr[0][S.index(',')+10:]
      Q = arr[1][5:]
      F = f'Formal form: For all x in D, if P(x) then Q(x)'
    else:
      D = S[10:S.index(',')]
      P = S[S.index(',')+10:S.rindex(',')]
      Q = S[S.rindex(',')+6:]
      F = f'Formal form: For all x in D, if P(x) then Q(x)'

  elif S.startswith('Exist a '):
    if S.__contains__('but'):
      arr = S.split('but')
      D = S[8:S.index(',')]
      if arr[0].__contains__('whose'):
        P = arr[0][S.index(',')+8:]
      else:
        P = arr[0][S.index(',')+6:]
      Q = arr[1]
      F = f'Formal form: Exist x in D, P(x) and Q(x)'

  elif S.startswith('Exist an '):
    D = S[9:S.index(',')]
    P = S[S.index(',')+6:S.rindex(',')]
    Q = S[S.rindex(',')+6:]
    F = f'Formal form: Exist x in D, P(x) and Q(x)'

  elif S.__contains__('did not '):
    arr = S.split(' did not ')
    D = arr[0].replace('Every ', '')
    P = "did not " + arr[1]
    Q = ""
    F = f'Formal form: For all x in D, P(x)'

  else:
    D = "Invalid Statement"
    P = "Invalid Statement"
    Q = "Invalid Statement"
    F = "Invalid Statement"

  return [D, P, Q, F]

S = [
  'For all people, if they are blond then they are westerners',
  'Exist a person, whose hair is black but is a westerner',
  'For all students, if they study correctly then they have high score',
  'For every mammal, if they live in the sea, they are either dolphins or whales',
  'For every bird, if they don’t have wings and can swim then they are penguins',
  'Exist a bird, who have wing but can’t fly'
]

for s in S:
    print(formalConvertPQ(s))
```
